```python
import time
import asyncio
import aiohttp
import nest_asyncio  # Para entornos como Jupyter Notebook

nest_asyncio.apply()  # Permite la reentrada en el bucle de eventos

# Función para realizar una solicitud asíncrona
async def fetch(session, url):
    async with session.get(url) as response:
        return await response.text()

# Función para probar una API
async def test_api(base_url, num_requests):
    async with aiohttp.ClientSession() as session:
        tasks = [fetch(session, f"{base_url}/get_value") for _ in range(num_requests)]
        results = await asyncio.gather(*tasks)
    return results

# Configuración de prueba
async def main():
    num_requests = 20

    # Prueba para Flask
    flask_start = time.time()
    try:
        await test_api("http://127.0.0.1:5000", num_requests)
    except Exception as e:
        print(f'Conexion Rechazada: {str(e)}')
    flask_end = time.time()
    print(f"Tiempo de ejecución Flask: {flask_end - flask_start:.2f} segundos.")

    # Prueba para FastAPI
    fastapi_start = time.time()
    try:
        await test_api("http://127.0.0.1:8000", num_requests)
    except Exception as e:
        print(f'Conexion Rechazada: {str(e)}')
    fastapi_end = time.time()
    print(f"Tiempo de ejecución FastAPI: {fastapi_end - fastapi_start:.2f} segundos.")

# Ejecutar el script principal directamente en Jupyter
await main()
```
