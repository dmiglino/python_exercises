## 01
Carga de datos de csv (valencia_pollution_dataset.csv) en un DataFrame

```python
# Voy a usar pandas para la carga del .csv directo en un DataFrame
path = os.path.join("", "valencia_pollution_dataset.csv")
df = pd.read_csv(path)

# Muestreo, no uso sample() porque me interesa ver mas un poco del head y un poco del tail
display(df)
display(df.info())
display(df.describe(include='all'))
```

## 02
- ¿Cuál es el rango temporal del dataset?
- Obtención del número de estaciones que tienen un sensor para medir monóxido de carbono

```python
# Uso esta funcion para formatear el datetime
def format_pd_timestamp(timestamp):
    return timestamp.strftime('%d-%m-%Y %X')
```

```python
# En primer lugar, voy a convertir la columna fecha_carga en un datetime
df['fecha_carga'] = pd.to_datetime(df['fecha_carga'])

min_dt = min(df['fecha_carga'])
max_dt = max(df['fecha_carga'])

print("Respuesta 2a:")
print(f'- El rango de fechas del dataset va desde {format_pd_timestamp(min_dt)} hasta {format_pd_timestamp(max_dt)}')
print(f"- Siendo el mismo un rango de {max_dt - min_dt}, o {pd.Timedelta(max_dt - min_dt).seconds} segundos.")
```

```python
# Obtención del número de estaciones que tienen un sensor para medir monóxido de carbono
mask = df['co'].notnull()
filas_con_sensor = df[mask]

# De ellas obtengo el listado de los 'objectid' para luego con .nunique() obtener el número de objectid's únicos (sin repetir)
estaciones_con_sensor = filas_con_sensor['objectid'].nunique()
print("Respuesta 2b:")
print(f"- Hay {estaciones_con_sensor} estaciones capaces de medir monóxido de carbono.")
print()

### Para hacerlo mas visual, podemos graficar la cantidad de estaciones con y sin sensor de CO.

# Obtengo la cantidad de estaciones que no poseen dicho sensor.
mask_nan = df['co'].isnull()

# De ellas, obtengo el listado de los 'objectid' para luego filtrar los duplicados
estaciones_sin_sensor = df[mask_nan]['objectid'].nunique()

# Valores para los ejes x e y respectivamente
labels = np.array(["Est. con Sensor", "Est. sin Sensor"])
valores = np.array([estaciones_con_sensor, estaciones_sin_sensor])

# Creo un grafico de torta 
fig = plt.figure(figsize=(10, 7))
plt.pie(valores, labels=labels, autopct='%1.0f%%', explode=(0,0.05), shadow=False,
        wedgeprops={"linewidth" : 1.3, "edgecolor": "grey"}, textprops={'size': 'large'})
plt.title('Cantidad de estaciones con o sin sensores para medir el CO')
plt.legend(labels=labels, loc='lower right')
plt.show()
```

## 03

- Número de estación con la media más alta de mediciones de monóxido de carbono

```python
# Interesante para una visualizacion rapida de las medias de todos los valores numericos agrupando por id de estacion
#print(df.groupby("objectid").mean(numeric_only=True))

# Para obtener la media de mediciones de CO, agrupo primero por id de estacion y luego obtengo la columna "co". 
# Con los registros agrupados y teniendo el "co", pido el "mean()" para obtener la media, los NaN son ignorados.
# Finalmente, con "idxmax()" obtengo el id de estacion con la media maxima, y con "max()" obtengo la media de esa estacion. 
media = df.groupby("objectid")['co'].mean()
id_estacion_media_max = media.idxmax()
valor_media_max = media.max()
print("Respuesta 3a:")
print(f'- La estación con id <{id_estacion_media_max}> registro la media máxima con un valor de <{valor_media_max}>')
print()

### Para hacerlo mas visual, podemos graficar la cantidad de estaciones con y sin sensor de CO.

# reset del tamaño de la figura, hay que hacerlo antes de crear el plot para que tome la configuración
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]

#filtro los NaN utilizando una mascara
media = media[media.notnull()]

# obtengo el listado de los objectid (id de estacion)
ids = list(media.index)

# ids es un objeto Index de pandas, por lo que matplotlib lo esta tomando como una serie de floats.
# voy a convertir cada uno a string, para graficar exactamente los id de estacion que quiero.
ids = [str(x) for x in ids]

# obtengo el listado de los valores de la media
med_values = list(media)

# armo los array con los valores para los ejes x e y
x = np.array(ids)
y = np.array(med_values)

# Creo un gráfico de barras horizontales con los valores de cada eje
plt.bar(x, y, color="rosybrown")
plt.title('Media de mediciones de monóxido de carbono')
plt.xlabel('Id de estación')
plt.ylabel('Media')
plt.grid(axis="x", linestyle='dashed', alpha=0.5)
#plt.figure(figsize=(8,4))
#plt.rcParams["figure.figsize"] = [8, 4]
#plt.figure(figsize=(14,7))
#plt.rcParams["figure.autolayout"] = True
plt.show()
```

## 04
Realizar el histograma de dióxido de nitrógeno de la estación del punto anterior

```python
# Obtengo primero los registros de la estación con la media máxima, y de ellos selecciono la columna "no2"
no2 = df.loc[df["objectid"]==id_estacion_media_max, "no2"]

### Estuve entendiendo el histograma y probando con diferentes bins. 
### Resulta interesante como ese parametro ayuda a una mejor visualización segun lo que se necesite.

# tamaño de la figura, hay que hacerlo antes de crear el plot para que tome la configuración
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
plt.rcParams["figure.figsize"] = [14, 7]

# Se me ocurrió entonces hacer 2 muestreos con diferentes bins, en 2 subplots dentro de la misma figura.
fig, ax = plt.subplots(1,2)

# El primero tiene un bins de 10 que es el valor por defecto. Esto significa que va a dividir el rango de valores en 10 bloques. 
# En este caso, al ser 27 el valor de medición maximo, lo divide por 10 y daría bloques de 2.7. 
ax[0].hist(no2, color='skyblue', edgecolor='black')

# Al segundo plot le voy a indicar que arme bloques de 2 en 2. 
#  Es decir, que agrupe el muestreo de valores de 0 a 2, de 2 a 4, de 4 a 6, etc.
ax[1].hist(no2, bins = np.arange(0, 30, 2), color='darkkhaki', edgecolor='black')

# Seteo de labels y titulo para cada subplot
ax[0].set_xlabel('Medición en 10 bloques/rangos')
ax[0].set_ylabel('Frequencia absoluta')
ax[1].set_xlabel('Medición con muestreo de 2 en 2')
ax[1].set_ylabel('Frequencia absoluta')
fig.suptitle('Histograma de Dióxido de Nitrogeno', fontsize=20)

plt.grid(axis="x", linestyle='dashed', alpha=0.5)
plt.show()
```

## 05
Realizar gráfica con  la línea temporal de las mediciones de ozono de la estación del punto anterior. Pista, la columna 'fecha_carga' habría que cambiarla de tipo.

```python
# Obtengo primero los registros de la estación con la media máxima, y de ellos selecciono la columna "o3"
ozono = df.loc[df["objectid"]==id_estacion_media_max, ["o3", "fecha_carga"]]
# Obtengo el nombre de la estación para usarla en el titulo del grafico
nombre_estacion = df.loc[df["objectid"]==id_estacion_media_max, 'nombre']

# tamaño de la figura, hay que hacerlo antes de crear el plot para que tome la configuración
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
plt.rcParams["figure.figsize"] = [14, 7]

# Creación de la linea temporal de mediciones
plt.plot(ozono["fecha_carga"], ozono["o3"], color='#5b3666', marker='o', linestyle='dashed', linewidth=1, markersize=6, label="O3")
plt.title(f"Mediciones de ozono de la estación {nombre_estacion[0]}", fontsize=13)
plt.xlabel("Fecha y hora de muestreo")
plt.ylabel("Medición de ozono")
plt.tick_params(axis='x', rotation=55)
plt.legend(loc='best')
plt.grid(linestyle='dashed', alpha=0.5)
plt.show()
```
